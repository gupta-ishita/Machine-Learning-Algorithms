```python
import numpy as np
```

```python
def gradient_descent(x, y):
  m_curr = b_curr = 1
  epochs = 10000
  n = len(x)
  learning_rate = 0.001

  for i in range(epochs):
    y_predicted = m_curr * x + b_curr
    cost = (1/n) * sum(np.square(y - y_predicted))
    m_deriv = -(2/n) * sum(x * (y - y_predicted))
    b_deriv = -(2/n) * sum(y - y_predicted)
    m_curr = m_curr - learning_rate * m_deriv
    b_curr = b_curr - learning_rate * b_deriv
    print("m {} , b {} , cost {} , epoch {}".format(m_curr, b_curr, cost, i))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
