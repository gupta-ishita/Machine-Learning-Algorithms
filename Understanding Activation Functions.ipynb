# **Activation Functions**
A neuron in the hidden layer mainly serves two purposes :

1) Calculates a value using linear relation between inputs and their corrsponding weights.

2) Applies activation function on the calculated value

The activation function provides non-linearity between the inputs and weight parameters. There are various types of activation functions.


**1) Step Function**

ANN_Unit_step.png

In step function, if the value is greater than a threshold then the neuron is fired up, in other words, neuron returns 1.

As this function returns only 0 or 1, it is not effective in multi-class classifier, wherein it might return 1 for more than 1 class causing confucsion.

Here is the implementation using python: 


```python
def step_function(inp, threshold):
  out = list(map(lambda ele:1 if ele >= threshold else 0, inp))
  return out    
```

```python
output = step_function([-1, 0, 2, 5], 1)
print(output)
```

```python
output = step_function([-1, -2, 0, 3, 2, -9], -1)
print(output)
```

**2) Sigmoid Function**


sig.png


Sigmoid takes a real value as the input and outputs another value between 0 and 1. It translates the input ranged in (-∞,∞) to the range in (0,1).
Advantages : 
It is differentiable at any point.
It maps the values of X in a smooth S-shaped curve and thus can be used in multi-class classifier.

Disadvantage: This function suffers from vanishing gradient. At either ends of the curve, the change in Y is almost negligible with chnages in X. This prevents the learning of the model during back propogation.

Sigmoid is still popular in classification problems. It is used in output layer.

The python code implementation :

```python
import math
def sigmoid(inp):
  out = 1 /(1 + math.exp(-inp))
  return out
```

```python
output = sigmoid(7)
print(output)
```

```python
output = sigmoid(-1)
print(output)
```

**3) Tanh Function**

tanh.png

The output values are bound in the range (-1, 1). 
The curve is non-linear and differentiable.

Tanh function also suffers from vanishing gradient.

The code implementation:

```python
import math
def tanh_activation_function(inp):
  out = (math.exp(inp) - math.exp(-inp)) / (math.exp(inp) + math.exp(-inp))
  return out
```

```python
output = tanh_activation_function(4)
print(output)
```

```python
output = tanh_activation_function(-4)
print(output)
```

**4) ReLU Function**

relu.png

The output values ranges between (0, inf).

It is a light weight function as it fires up only for positive values

ReLU also has vanishing gradient problem for values < 0

Below is the code implementation in python :

```python
def relu_function(inp):
  return max(0, inp)
```

```python
output = relu_function(4)
print(output)
```

```python
output = relu_function(-4)
print(output)
```

**5) Leaky ReLU Function**

leaky Relu.png


The code implementation for the same is below:

```python
def leaky_relu(inp):
  return max(0.1*inp, inp)
```

```python
output = leaky_relu(10)
print(output)
```

```python
output = leaky_relu(-10)
print(output)
```

**6) Softmax Function**

softmax.jpeg


The code implementation for the same is below:

```python
from numpy import exp
def softmax_function(inp):
  out = exp(inp)
  return out / out.sum()
```

```python
output = softmax_function([1.3,5.1,2.2,0.7,1.1])
print(output)
```

```python
output = softmax_function([1,3,2,5])
print(output)
```
